# metinler.txt dosyasındaki metinleri index ile birlikte cümleleri satırlara yazdırma


```python
import csv
import nltk
from nltk.tokenize import sent_tokenize

def cumleleri_birlestir_ve_kaydet(dosya_adi):
    try:
        with open(dosya_adi, 'r', encoding='utf-8') as dosya:
            metin = dosya.read()

            # Cümleleri ayır
            cumleler = sent_tokenize(metin)

            # Cümleleri birleştir ve indeksle
            birlesik_cumleler = []
            indeks = 1

            for cumle in cumleler:
                # Alt satıra geçmiş bölümleri birleştir
                cumle_birlestirilmis = cumle.replace('\n', ' ')
                birlesik_cumleler.append((indeks, cumle_birlestirilmis.strip()))
                indeks += 1

            # CSV dosyasına yaz
            csv_dosya_adi = "derlem.csv"
            with open(csv_dosya_adi, 'w', newline='', encoding='utf-8') as csvdosya:
                yazici = csv.writer(csvdosya)
                yazici.writerow(['Index', 'Cumle'])

                for indeks, cumle in birlesik_cumleler:
                    yazici.writerow([indeks, cumle])

            print("Cümleler CSV dosyasına kaydedildi: {}".format(csv_dosya_adi))

    except FileNotFoundError:
        print("Dosya bulunamadı: {}".format(dosya_adi))
    except Exception as e:
        print("Bir hata oluştu: {}".format(str(e)))


dosya_adi = "metinler.txt"  
cumleleri_birlestir_ve_kaydet(dosya_adi)
```

# Stop words, nümerik ifadeler ve özel karakterleri ve noktalama işaretlerini metinden kaldırma

```python
import re
import string
import pandas as pd

def numerik_ifadeleri_sil(metin):
    numerik_ifadeler = re.findall(r'\d+', metin)
    for ifade in numerik_ifadeler:
        metin = metin.replace(ifade, '')
    return metin

def ozel_karakter_noktalama_sil(metin):
    ozel_karakterler_noktalama = re.findall('[^A-Za-z0-9ğüşıöçĞÜŞİÖÇ\s]', metin)
    for ifade in ozel_karakterler_noktalama:
        metin = metin.replace(ifade, '')
    return metin

def stopwordleri_sil(metin):
    
    stopword_liste = ["acaba", "ama", "ancak", "artık", "asla", "aslında", "az", "bazen", "bazı", "bazıları", "bile", "bir", "birçok", "birçoğu", "biri", "birisi", "birkaç", "birkaçı", "birşey", "birşeyi", "biz", "bizden", "bize", "bizden", "bizim", "böyle", "böylece", "bu", "buna", "bunda", "bundan", "bunu", "bunun", "burada", "çok", "çünkü", "da", "daha", "de", "defa", "diğer", "diye", "dolayı", "eğer", "en", "gibi", "hem", "hep", "hepsi", "her", "herhangi", "herkes", "herşey", "hiç", "hiçbir", "için", "ile", "ise", "işte", "kaç", "kadar", "kendi", "kim", "kimden", "kime", "kimi", "kimi", "kimisi", "madem", "mi", "mı", "mu", "mü", "nasıl", "ne", "ne kadar", "ne zaman", "neden", "nerde", "nerede", "nereden", "nereye", "nereye", "nasıl", "nasıl", "niçin", "niçin", "o", "olan", "olarak", "oldu", "olduğu", "olduğunu", "olmadı", "olmaz", "olmayan", "olmuyor", "olsa", "olsun", "olup", "olur", "olursa", "oluyor", "on", "ona", "ondan", "onlar", "onlardan", "onlari", "onlarla", "onların", "onu", "onun", "orada", "o zaman", "sana", "sen", "senden", "seni", "senin", "şey", "şeyden", "şeyi", "şeylere", "şeyler", "şu", "şuna", "şunda", "şundan", "şunlar", "şunu", "şunun", "tarafından", "tüm", "ve", "veya", "ya", "ya da", "yani", "yerine", "yetmiş", "yetmişer", "yetmişer", "yetmişi", "yine", "yirmi", "yirmi", "yirmi", "yirmisi", "yoksa", "yüz", "yüzde"]
    
    kelimeler = metin.split()
    temizlenmis_metin = ' '.join([kelime for kelime in kelimeler if kelime.lower() not in map(str.lower, stopword_liste)])
    
    return temizlenmis_metin

def kucuk_harfe_cevir(metin):
    return metin.lower()

# "derlem.csv" dosyasını okuma
df = pd.read_csv("derlem.csv", encoding="utf-8")

# Metni küçük harfe çevirme işlemi
df['kucukharf_sonuc'] = df['Cumle'].apply(kucuk_harfe_cevir)

# Numerik ifadelerden arındırma işlemi
df['numerik_ifadeler_sonuc'] = df['Cumle'].apply(numerik_ifadeleri_sil)

# Özel karakterlerden ve noktalama işaretlerinden arındırma işlemi
df['ozel_karakter_noktalama_sonuc'] = df['Cumle'].apply(ozel_karakter_noktalama_sil)

# Stopword'leri kaldırma işlemi
df['temizlenmis_cümle'] = df['Cumle'].apply(stopwordleri_sil)


# Sonucu CSV dosyalarına yazma
df[['Index', 'numerik_ifadeler_sonuc']].to_csv("numerik_ifadeler_sonuc.csv", index=False, encoding="utf-8")
df[['Index', 'ozel_karakter_noktalama_sonuc']].to_csv("ozel_karakter_noktalama_sonuc.csv", index=False, encoding="utf-8")
df[['Index', 'temizlenmis_cümle']].to_csv("derlem_stopword_sonuc.csv", index=False, encoding="utf-8")
df[['Index', 'kucukharf_sonuc']].to_csv("derlem_kucukharf_sonuc.csv", index=False, encoding="utf-8")
```
